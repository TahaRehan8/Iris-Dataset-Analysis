#**Plotting Graphs For relationships**



```python
fig,sub=plt.subplots(1,2,figsize=(10,4))

sb.scatterplot(x='petal_length',y='petal_width',data=dataset,hue='species',ax=sub[0])
sub[0].set_title("Petal Length vs Petal Width by Species")
sb.scatterplot(x='sepal_length',y='sepal_width',data=dataset,hue='species',ax=sub[1])
sub[1].set_title("Sepal Length vs Sepal Width by Species")
plt.tight_layout()
plt.show()
```

```python
fig,sub=plt.subplots(1,2,figsize=(10,4))
sb.scatterplot(x='sepal_length',y='petal_length',data=dataset,hue='species',ax=sub[0])
sub[0].set_title("Sepal Length vs Petal length by Species")
sb.scatterplot(x='sepal_width',y='petal_width',data=dataset,hue='species',ax=sub[1])
sub[1].set_title("Sepal Width vs Petal Width by Species")
plt.tight_layout()
plt.show()
```

#**Histogram**

```python
colms=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Create 2x2 subplot grid
fig,axes = plt.subplots(2, 2, figsize=(10, 5))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot histograms for each column
for i, col in enumerate(colms):
    sb.histplot(data=dataset, x=col, ax=axes[i], bins=20)
    axes[i].set_title(f'Histogram of {col}')

plt.tight_layout()
plt.show()
```

#**Box Plotting**

```python
colms = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Create subplots: 2 rows, 2 columns
fig, axes = plt.subplots(2, 2, figsize=(10, 5))
axes = axes.flatten()

for i, col in enumerate(colms):
    sb.boxplot(data=dataset, y=col, ax=axes[i])
    axes[i].set_title(f'Box Plot of {col}')

plt.tight_layout()
plt.show()
```
